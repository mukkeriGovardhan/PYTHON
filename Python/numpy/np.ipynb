ARANGE

```python
np.arange() # it will print the array of values from 0 to 9 because the default start value is 0 and the default step value is 1 and the end value is 10 but not include 10
```

```python
np.arange(10)  #arange it will print the value within interval 
```

```python
np.arange(10,15) # it will print the value from 10 to 15 but not include 15
```

```python
np.arange(10,20,5) # it will print the value from 10 to 20 but not include 20 and the step size is 5
```

```python
np.arange(10,20,5,2) # it will print the value from 10 to 20 but not include 20 and the step size is 5 and the data type is int16
```

```python
np.arange(10,20) # it will print the value from 10 to 20 but not include 20 and the step size is 1 and the data type is int64
```

```python
np.arange(20, 10) # it will print the value from 20 to 10 but not include 10 and the step size is 1 and the data type is int64
```

```python
np.arange(-20,10) # it will print the value from -20 to 10 but not include 10 and the step size is 1 and the data type is int64
```
