DATATYPES


```python
#NUMERIC
import sys
val1 = 10
print(val1)
print(type(val1))#types of data
print(sys.getsizeof(val1))#size of integer object  in bytes
print(val1, "is Integer:", isinstance(val1, int))#check if val1 is an integer

```

```python
val2 = 3.14 #float data type
print(val2)
print(type(val2)) #types of data
print(sys.getsizeof(val2)) #size of float object in bytes
print(val2, "is Float:", isinstance(val2, float)) #check if val2 is a float
```

```python
val3 = 2 + 3j #complex data type
print(val3) 
print(type(val3)) #types of data
print(sys.getsizeof(val3)) #size of complex object in bytes
print(val3, "is Complex:", isinstance(val3, complex)) #check if val3 is a complex number
```
