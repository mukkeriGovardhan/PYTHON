Varible assigment


```python
intvar = 10
floatvar = 3.14
strvar = "Hello, World!"
boolvar = True
print(intvar) #This will print the integer variable intvar, which has the value 10
print(floatvar) #This will print the float variable floatvar, which has the value 3.14
print(strvar) #This will print the string variable strvar, which has the value "Hello, World!"
print(boolvar) #This will print the boolean variable boolvar, which has the value True
```
