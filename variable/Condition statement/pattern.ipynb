# right angle triangle pattern 


```python
for i in range(1,6):
  print('*' *i)
```

# inverted right angle triangle pattern


```python
for i in range(5,0,-1):
    print('*' *i)
```

# pyramid pattern 

```python
for i in range (1,6):
    print(' '*(6-i)+ '*' *(2*i-1))
```

```python
for i in range (6,0,-1):
    print(' '*(6-i)+ '*' *(2*i-1))
```

# Diamond pattern 

```python
for i in range (1,6):
   print(' '*(5-i)+ '*' *(2*i-1))
for i in range(4,0,-1):
    print(' '*(5-i)+ '*' *(2*i-1)) 
```

# hallow square pattern

```python
for i in range(5):
    for j in range(5):
        if i ==0 or i ==4 or j ==0 or j == 4:
            print('*',end= '')
        else:
            print(' ',end='')
    print()
```

# full square pattern

```python
for i in range(5):
    print(' * ' *5)
```

# right angle triangle(number pattern)

```python
for i in range(1,6):
    print(' '.join(str(x) for x in range(1,i+1)))
```

```python
for i in range(6,0,-1):
    print(' '.join(str(x) for x in range(1,i+1)))
```

# floyd triangle

```python
nums = 1
for i in range(1,6):
    for j in range(1,i+1):
        print(nums,end= ' ')
        nums+=1
    print()
```
